## 🤔 학습 전 생각해보기

현실 세계에서도 우리는 구분합니다:
- 종이에 적힌 "3" → 그림 (문자)
- 사과 3개 → 실제 개수 (숫자)

컴퓨터도 이걸 구분해야 계산을 할 수 있어요!

---

## 💡 핵심 개념 미리보기

```python
# 자료형의 차이
text_10 = "10"      # 문자열 (str)
number_10 = 10      # 정수 (int)

# 결과가 다릅니다!
print(text_10 + text_10)     # "1010" (문자 연결)
print(number_10 + number_10)  # 20 (숫자 계산)

# 형변환으로 해결
result = int(text_10) + int(text_10)  # 20
```

## 🎮 미리보기: 이 파트가 끝나면 만들 수 있는 것

```python
# 제대로 작동하는 계산기
num1 = float(input("첫 번째 숫자: "))  # 형변환!
num2 = float(input("두 번째 숫자: "))

print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} × {num2} = {num1 * num2}")

# BMI 계산기
height = float(input("키(cm): ")) / 100  # cm를 m로
weight = float(input("몸무게(kg): "))
bmi = weight / (height ** 2)
print(f"당신의 BMI: {bmi:.1f}")
```
