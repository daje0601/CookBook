## 6ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: í•™ìƒ ì •ë³´ ê´€ë¦¬

```python
# ğŸ® ì‹¤ìŠµ: íŠœí”Œë¡œ í•™ìƒ ì •ë³´ ê´€ë¦¬
print("ğŸ“ í•™ìƒ ì •ë³´ ê´€ë¦¬ ì‹œìŠ¤í…œ")
print("=" * 40)

# í•™ìƒ ì •ë³´ëŠ” íŠœí”Œë¡œ (ë³€ê²½ ë°©ì§€)
students = [
    ("20240001", "ê¹€ì² ìˆ˜", 85, 90, 88),  # (í•™ë²ˆ, ì´ë¦„, êµ­ì–´, ì˜ì–´, ìˆ˜í•™)
    ("20240002", "ì´ì˜í¬", 92, 88, 95),
    ("20240003", "ë°•ë¯¼ìˆ˜", 78, 85, 82),
]

def display_student(student):
    student_id, name, *scores = student  # ì–¸íŒ¨í‚¹
    avg = sum(scores) / len(scores)
    
    print(f"í•™ë²ˆ: {student_id}")
    print(f"ì´ë¦„: {name}")
    print(f"ì ìˆ˜: êµ­ì–´={scores[0]}, ì˜ì–´={scores[1]}, ìˆ˜í•™={scores[2]}")
    print(f"í‰ê· : {avg:.1f}ì ")
    print("-" * 30)

# ëª¨ë“  í•™ìƒ ì •ë³´ ì¶œë ¥
for student in students:
    display_student(student)

# íŠœí”Œì´ë¯€ë¡œ ì‹¤ìˆ˜ë¡œ ìˆ˜ì •ë  ìœ„í—˜ì´ ì—†ìŒ!
# students[0][2] = 100  # TypeError!
```

## 7ï¸âƒ£ íŠœí”Œê³¼ í•¨ìˆ˜ ë°˜í™˜ê°’

```python
# ğŸ® ì‹¤ìŠµ: ì—¬ëŸ¬ ê°’ ë°˜í™˜í•˜ê¸°

def calculate_rectangle(width, height):
    area = width * height
    perimeter = 2 * (width + height)
    diagonal = (width**2 + height**2) ** 0.5
    
    # íŠœí”Œë¡œ ì—¬ëŸ¬ ê°’ ë°˜í™˜
    return area, perimeter, diagonal

# ë°˜í™˜ê°’ ë°›ê¸°
result = calculate_rectangle(3, 4)
print(f"ê²°ê³¼ (íŠœí”Œ): {result}")

# ì–¸íŒ¨í‚¹ìœ¼ë¡œ ê°ê° ë°›ê¸°
area, perimeter, diagonal = calculate_rectangle(3, 4)
print(f"ë©´ì : {area}")
print(f"ë‘˜ë ˆ: {perimeter}")
print(f"ëŒ€ê°ì„ : {diagonal:.2f}")
```
