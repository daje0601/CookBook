```python
# ğŸ® ì‹¤ìŠµ: ìš”ì†Œ ì¶”ê°€
fruits = {"ì‚¬ê³¼", "ë°”ë‚˜ë‚˜"}
print(f"ì´ˆê¸°: {fruits}")

# add(): í•œ ê°œ ì¶”ê°€
fruits.add("ì˜¤ë Œì§€")
print(f"add í›„: {fruits}")

# update(): ì—¬ëŸ¬ ê°œ ì¶”ê°€
fruits.update(["í¬ë„", "ìˆ˜ë°•"])
print(f"update í›„: {fruits}")

# ì¤‘ë³µ ì¶”ê°€ ì‹œë„
fruits.add("ì‚¬ê³¼")  # ì´ë¯¸ ìˆìŒ
print(f"ì¤‘ë³µ ì¶”ê°€: {fruits}")  # ë³€í™” ì—†ìŒ
```

```python
# ğŸ® ì‹¤ìŠµ: ìš”ì†Œ ì‚­ì œ
numbers = {1, 2, 3, 4, 5}
print(f"ì´ˆê¸°: {numbers}")

# remove(): íŠ¹ì • ìš”ì†Œ ì‚­ì œ (ì—†ìœ¼ë©´ ì—ëŸ¬)
numbers.remove(3)
print(f"remove(3) í›„: {numbers}")

# discard(): íŠ¹ì • ìš”ì†Œ ì‚­ì œ (ì—†ì–´ë„ OK)
numbers.discard(10)  # ì—†ì§€ë§Œ ì—ëŸ¬ ì•ˆ ë‚¨
print(f"discard(10) í›„: {numbers}")

# pop(): ì„ì˜ì˜ ìš”ì†Œ ì‚­ì œí•˜ê³  ë°˜í™˜
removed = numbers.pop()
print(f"pop() í›„: {numbers}, ì‚­ì œëœ ê°’: {removed}")

# clear(): ëª¨ë‘ ì‚­ì œ
numbers.clear()
print(f"clear() í›„: {numbers}")
```

## 3ï¸âƒ£ ì§‘í•© ì—°ì‚° - ìˆ˜í•™ì  ì§‘í•©

```python
# ğŸ® ì‹¤ìŠµ: í•©ì§‘í•©ê³¼ êµì§‘í•©
python_users = {"ì² ìˆ˜", "ì˜í¬", "ë¯¼ìˆ˜", "ì§€í›ˆ"}
java_users = {"ì˜í¬", "ë¯¼ìˆ˜", "í˜„ì •", "ìˆ˜ì§„"}

print(f"Python ì‚¬ìš©ì: {python_users}")
print(f"Java ì‚¬ìš©ì: {java_users}")

# í•©ì§‘í•© (Union): ëª¨ë“  ì‚¬ìš©ì
all_users = python_users | java_users  # ë˜ëŠ” .union()
print(f"\nëª¨ë“  ì‚¬ìš©ì: {all_users}")

# êµì§‘í•© (Intersection): ë‘ ì–¸ì–´ ëª¨ë‘ ì‚¬ìš©
both_users = python_users & java_users  # ë˜ëŠ” .intersection()
print(f"ë‘ ì–¸ì–´ ëª¨ë‘: {both_users}")
```

```python
# ğŸ® ì‹¤ìŠµ: ì°¨ì§‘í•©ê³¼ ëŒ€ì¹­ì°¨ì§‘í•©

# ì°¨ì§‘í•© (Difference): Pythonë§Œ ì‚¬ìš©
python_only = python_users - java_users  # ë˜ëŠ” .difference()
print(f"Pythonë§Œ: {python_only}")

# Javaë§Œ ì‚¬ìš©
java_only = java_users - python_users
print(f"Javaë§Œ: {java_only}")

# ëŒ€ì¹­ì°¨ì§‘í•© (Symmetric Difference): í•œ ì–¸ì–´ë§Œ ì‚¬ìš©
exclusive_users = python_users ^ java_users  # ë˜ëŠ” .symmetric_difference()
print(f"\ní•œ ì–¸ì–´ë§Œ: {exclusive_users}")
```

## 6ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: ì¹œêµ¬ ì¶”ì²œ ì‹œìŠ¤í…œ

```python
# ğŸ® ì‹¤ìŠµ: SNS ì¹œêµ¬ ì¶”ì²œ
print("ğŸ‘¥ ì¹œêµ¬ ì¶”ì²œ ì‹œìŠ¤í…œ")
print("=" * 40)

# ê° ì‚¬ìš©ìì˜ ì¹œêµ¬ ëª©ë¡
friends = {
    "ì² ìˆ˜": {"ì˜í¬", "ë¯¼ìˆ˜", "ì§€í›ˆ", "í˜„ì •"},
    "ì˜í¬": {"ì² ìˆ˜", "ë¯¼ìˆ˜", "ìˆ˜ì§„"},
    "ë¯¼ìˆ˜": {"ì² ìˆ˜", "ì˜í¬", "ì§€í›ˆ"},
}

def recommend_friends(user_name, friends_dict):
    if user_name not in friends_dict:
        return set()
    
    my_friends = friends_dict[user_name]
    recommendations = set()
    
    # ì¹œêµ¬ì˜ ì¹œêµ¬ íƒìƒ‰
    for friend in my_friends:
        if friend in friends_dict:
            # ì¹œêµ¬ì˜ ì¹œêµ¬ë“¤ ì¶”ê°€
            recommendations |= friends_dict[friend]
    
    # ìê¸° ìì‹ ê³¼ ì´ë¯¸ ì¹œêµ¬ì¸ ì‚¬ëŒ ì œì™¸
    recommendations -= {user_name}
    recommendations -= my_friends
    
    return recommendations

# ê° ì‚¬ìš©ìì—ê²Œ ì¹œêµ¬ ì¶”ì²œ
for user in friends:
    rec = recommend_friends(user, friends)
    print(f"{user}ì˜ ì¹œêµ¬: {friends[user]}")
    if rec:
        print(f"  â¡ï¸ ì¶”ì²œ: {rec}")
    else:
        print(f"  â¡ï¸ ì¶”ì²œ ì—†ìŒ")
    print()
```
