## 7ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„

```python
# ğŸ® ì‹¤ìŠµ: ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„
import random

def rock_paper_scissors():
    choices = ["ê°€ìœ„", "ë°”ìœ„", "ë³´"]
    
    print("âœ‚ï¸ğŸ“„ğŸ—¿ ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„!")
    print("=" * 30)
    
    # ì‚¬ìš©ì ì…ë ¥
    print("ì„ íƒì§€: ê°€ìœ„, ë°”ìœ„, ë³´")
    user_choice = input("ë‹¹ì‹ ì˜ ì„ íƒ: ").strip()
    
    # ì…ë ¥ ê²€ì¦
    if user_choice not in choices:
        print("âŒ ì˜¬ë°”ë¥¸ ì„ íƒì„ ì…ë ¥í•´ì£¼ì„¸ìš”!")
        return
    
    # ì»´í“¨í„° ì„ íƒ
    computer_choice = random.choice(choices)
    
    print(f"\në‹¹ì‹ : {user_choice}")
    print(f"ì»´í“¨í„°: {computer_choice}")
    print()
    
    # ê²°ê³¼ íŒì •
    if user_choice == computer_choice:
        result = "ë¬´ìŠ¹ë¶€"
        emoji = "ğŸ˜"
        
    elif (
        (user_choice == "ê°€ìœ„" and computer_choice == "ë³´") or
        (user_choice == "ë°”ìœ„" and computer_choice == "ê°€ìœ„") or
        (user_choice == "ë³´" and computer_choice == "ë°”ìœ„")
    ):
        result = "ìŠ¹ë¦¬"
        emoji = "ğŸ‰"
        
    else:
        result = "íŒ¨ë°°"
        emoji = "ğŸ˜¢"
    
    # ê²°ê³¼ ì¶œë ¥
    print("=" * 30)
    print(f"ê²°ê³¼: {result} {emoji}")
    
    if result == "ìŠ¹ë¦¬":
        print("ì¶•í•˜í•©ë‹ˆë‹¤! ğŸ†")
    elif result == "íŒ¨ë°°":
        print("ë‹¤ìŒ ê¸°íšŒì—! ğŸ’ª")
    else:
        print("í•œ ë²ˆ ë” í•´ë³¼ê¹Œìš”? ğŸ”„")

# ê²Œì„ ì‹¤í–‰
rock_paper_scissors()
```

## 8ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: BMI ê³„ì‚°ê¸°

```python
# ğŸ® ì‹¤ìŠµ: BMI ê³„ì‚° ë° íŒì •

def calculate_bmi():
    print("ğŸ“Š BMI ê³„ì‚°ê¸°")
    print("=" * 20)
    
    try:
        # ì‚¬ìš©ì ì…ë ¥
        height = float(input("í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš” (cm): "))
        weight = float(input("ëª¸ë¬´ê²Œë¥¼ ì…ë ¥í•˜ì„¸ìš” (kg): "))
        
        # ì…ë ¥ê°’ ê²€ì¦
        if height <= 0 or weight <= 0:
            print("âŒ í‚¤ì™€ ëª¸ë¬´ê²ŒëŠ” 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤!")
            return
            
        if height > 10:  # cm ë‹¨ìœ„ë¡œ ê°€ì •
            height_m = height / 100
        else:  # ì´ë¯¸ m ë‹¨ìœ„
            height_m = height
        
        # BMI ê³„ì‚°
        bmi = weight / (height_m ** 2)
        
        print(f"\nğŸ“ í‚¤: {height}cm")
        print(f"âš–ï¸ ëª¸ë¬´ê²Œ: {weight}kg")
        print(f"ğŸ“Š BMI: {bmi:.1f}")
        
        # BMI íŒì •
        if bmi < 18.5:
            category = "ì €ì²´ì¤‘"
            emoji = "ğŸ“‰"
            advice = "ì˜ì–‘ ì„­ì·¨ë¥¼ ëŠ˜ë ¤ë³´ì„¸ìš”."
            
        elif bmi < 23:  # í•œêµ­ ê¸°ì¤€
            category = "ì •ìƒì²´ì¤‘"
            emoji = "âœ…"
            advice = "ê±´ê°•í•œ ì²´ì¤‘ì„ ìœ ì§€í•˜ê³  ê³„ì‹œë„¤ìš”!"
            
        elif bmi < 25:
            category = "ê³¼ì²´ì¤‘"
            emoji = "âš ï¸"
            advice = "ê°€ë²¼ìš´ ìš´ë™ì„ ì‹œì‘í•´ë³´ì„¸ìš”."
            
        elif bmi < 30:
            category = "ë¹„ë§Œ"
            emoji = "ğŸš¨"
            advice = "ê·œì¹™ì ì¸ ìš´ë™ê³¼ ì‹ë‹¨ ê´€ë¦¬ê°€ í•„ìš”í•©ë‹ˆë‹¤."
            
        else:
            category = "ê³ ë„ë¹„ë§Œ"
            emoji = "ğŸ†˜"
            advice = "ì „ë¬¸ì˜ì™€ ìƒë‹´ì„ ë°›ì•„ë³´ì„¸ìš”."
        
        # ê²°ê³¼ ì¶œë ¥
        print("\n" + "=" * 30)
        print(f"íŒì •: {category} {emoji}")
        print(f"ì¡°ì–¸: {advice}")
        
        # ì¶”ê°€ ì •ë³´
        if category != "ì •ìƒì²´ì¤‘":
            if bmi < 18.5:
                target_weight = 18.5 * (height_m ** 2)
                print(f"ê¶Œì¥ ìµœì†Œ ì²´ì¤‘: {target_weight:.1f}kg (+{target_weight - weight:.1f}kg)")
            elif bmi >= 23:
                target_weight = 22.5 * (height_m ** 2)  # ì •ìƒ ë²”ìœ„ ì¤‘ê°„ê°’
                print(f"ê¶Œì¥ ì²´ì¤‘: {target_weight:.1f}kg ({target_weight - weight:+.1f}kg)")
    
    except ValueError:
        print("âŒ ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!")
    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")

# BMI ê³„ì‚°ê¸° ì‹¤í–‰
calculate_bmi()
```

## ğŸ”Ÿ ì‹¤ì „ ì˜ˆì œ: ê°„ë‹¨í•œ í€´ì¦ˆ ê²Œì„

```python
# ğŸ® ì‹¤ìŠµ: ë‹¤ì¤‘ ì„ íƒ í€´ì¦ˆ

def quiz_game():
    print("ğŸ§  íŒŒì´ì¬ í€´ì¦ˆ ê²Œì„!")
    print("=" * 30)
    
    questions = [
        {
            "question": "Pythonì˜ ì°½ì‹œìëŠ” ëˆ„êµ¬ì¸ê°€ìš”?",
            "options": ["1. ê·€ë„ ë°˜ ë¡œì„¬", "2. ë¹Œ ê²Œì´ì¸ ", "3. ìŠ¤í‹°ë¸Œ ì¡ìŠ¤", "4. ë§ˆí¬ ì €ì»¤ë²„ê·¸"],
            "answer": 1
        },
        {
            "question": "Pythonì—ì„œ ë¦¬ìŠ¤íŠ¸ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê¸°í˜¸ëŠ”?",
            "options": ["1. {}", "2. []", "3. ()", "4. <>"],
            "answer": 2
        },
        {
            "question": "print('Hello') ì˜ ê²°ê³¼ëŠ”?",
            "options": ["1. 'Hello'", "2. Hello", "3. \"Hello\"", "4. ì˜¤ë¥˜"],
            "answer": 2
        }
    ]
    
    score = 0
    
    for i, q in enumerate(questions, 1):
        print(f"\në¬¸ì œ {i}: {q['question']}")
        for option in q['options']:
            print(f"  {option}")
        
        try:
            user_answer = int(input("ë‹µ (1-4): "))
            
            if user_answer < 1 or user_answer > 4:
                print("âŒ 1-4 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”!")
                continue
                
            if user_answer == q['answer']:
                print("âœ… ì •ë‹µ! ğŸ‰")
                score += 1
            else:
                print(f"âŒ í‹€ë ¸ìŠµë‹ˆë‹¤. ì •ë‹µì€ {q['answer']}ë²ˆì…ë‹ˆë‹¤.")
                
        except ValueError:
            print("âŒ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!")
    
    # ê²°ê³¼ ì¶œë ¥
    print("\n" + "=" * 30)
    print(f"ë‹¹ì‹ ì˜ ì ìˆ˜: {score}/{len(questions)}")
    
    percentage = (score / len(questions)) * 100
    
    if percentage == 100:
        grade = "ì™„ë²½"
        emoji = "ğŸ†"
    elif percentage >= 80:
        grade = "ìš°ìˆ˜"
        emoji = "ğŸ˜Š"
    elif percentage >= 60:
        grade = "ë³´í†µ"
        emoji = "ğŸ˜"
    else:
        grade = "ë¶„ë°œ"
        emoji = "ğŸ’ª"
    
    print(f"í‰ê°€: {grade} {emoji} ({percentage:.0f}%)")

# í€´ì¦ˆ ê²Œì„ ì‹¤í–‰
quiz_game()
```
