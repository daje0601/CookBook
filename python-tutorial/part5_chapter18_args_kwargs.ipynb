## 1ï¸âƒ£ *args - ê°€ë³€ ìœ„ì¹˜ ì¸ì

```python
# ğŸ® ì‹¤ìŠµ: *args ê¸°ë³¸

def print_args(*args):
    print(f"argsì˜ íƒ€ì…: {type(args)}")  # íŠœí”Œ!
    print(f"argsì˜ ë‚´ìš©: {args}")
    print(f"ê°œìˆ˜: {len(args)}ê°œ")
    
    # í•˜ë‚˜ì”© ì²˜ë¦¬
    for i, arg in enumerate(args, 1):
        print(f"  {i}ë²ˆì§¸ ì¸ì: {arg}")

# ë‹¤ì–‘í•œ í˜¸ì¶œ
print("ğŸ”¸ ì¸ì 1ê°œ:")
print_args("Hello")

print("\nğŸ”¸ ì¸ì 3ê°œ:")
print_args(10, 20, 30)

print("\nğŸ”¸ ì¸ì 0ê°œ:")
print_args()
```

```python
# ğŸ® ì‹¤ìŠµ: ì‹¤ìš©ì ì¸ *args í™œìš©

def calculate_average(*scores):
    """ì—¬ëŸ¬ ì ìˆ˜ì˜ í‰ê·  ê³„ì‚°"""
    if not scores:
        return 0
    
    total = sum(scores)
    count = len(scores)
    average = total / count
    
    print(f"ğŸ“Š ì„±ì  ë¶„ì„")
    print(f"  ì ìˆ˜: {scores}")
    print(f"  í•©ê³„: {total}ì ")
    print(f"  ê°œìˆ˜: {count}ê°œ")
    print(f"  í‰ê· : {average:.1f}ì ")
    
    return average

# ì‚¬ìš© ì˜ˆì‹œ
avg1 = calculate_average(85, 90, 78)
print()
avg2 = calculate_average(90, 95, 88, 92, 85)
print()
avg3 = calculate_average(100)  # í•˜ë‚˜ë§Œì´ë¼ë„ OK
```

## 2ï¸âƒ£ **kwargs - ê°€ë³€ í‚¤ì›Œë“œ ì¸ì

```python
# ğŸ® ì‹¤ìŠµ: **kwargs ê¸°ë³¸

def print_kwargs(**kwargs):
    print(f"kwargsì˜ íƒ€ì…: {type(kwargs)}")  # ë”•ì…”ë„ˆë¦¬!
    print(f"kwargsì˜ ë‚´ìš©: {kwargs}")
    print(f"ê°œìˆ˜: {len(kwargs)}ê°œ")
    
    # í‚¤-ê°’ ìŒ ì²˜ë¦¬
    for key, value in kwargs.items():
        print(f"  {key} = {value}")

# ë‹¤ì–‘í•œ í˜¸ì¶œ
print("ğŸ”¹ í‚¤ì›Œë“œ ì¸ì 2ê°œ:")
print_kwargs(name="ì² ìˆ˜", age=25)

print("\nğŸ”¹ í‚¤ì›Œë“œ ì¸ì 4ê°œ:")
print_kwargs(city="ì„œìš¸", country="í•œêµ­", language="Python", level="Junior")

print("\nğŸ”¹ í‚¤ì›Œë“œ ì¸ì 0ê°œ:")
print_kwargs()
```

```python
# ğŸ® ì‹¤ìŠµ: ì‹¤ìš©ì ì¸ **kwargs í™œìš©

def create_profile(name, **details):
    """ì‚¬ìš©ì í”„ë¡œí•„ ìƒì„±"""
    profile = {'name': name}
    profile.update(details)  # kwargsë¥¼ profileì— ì¶”ê°€
    
    print(f"ğŸ‘¤ {name}ë‹˜ì˜ í”„ë¡œí•„")
    print("-" * 30)
    
    for key, value in profile.items():
        # í‚¤ ì´ë¦„ì„ ë³´ê¸° ì¢‹ê²Œ ë³€í™˜
        display_key = key.replace('_', ' ').title()
        print(f"{display_key}: {value}")
    
    return profile

# ì‚¬ìš© ì˜ˆì‹œ
profile1 = create_profile("이ì² ìˆ˜", age=25, job="ê°œë°œì", city="ì„œìš¸")
print()

profile2 = create_profile(
    "ê¹€ì˜í¬",
    age=28,
    job="ë””ìì´ë„ˆ",
    city="ë¶€ì‚°",
    hobby="ê·¸ë¦¼",
    favorite_color="íŒŒë‘"
)
```

## 3ï¸âƒ£ *argsì™€ **kwargs í•¨ê»˜ ì‚¬ìš©í•˜ê¸°

```python
# ğŸ® ì‹¤ìŠµ: ëª¨ë“  ë§¤ê°œë³€ìˆ˜ íƒ€ì… í•¨ê»˜ ì‚¬ìš©

def super_function(required, default="default", *args, **kwargs):
    """ëª¨ë“  íƒ€ì…ì˜ ë§¤ê°œë³€ìˆ˜ë¥¼ ë°›ëŠ” í•¨ìˆ˜"""
    print("ğŸŒŸ ìˆ˜í¼ í•¨ìˆ˜ í˜¸ì¶œ")
    print("-" * 40)
    print(f"í•„ìˆ˜ ë§¤ê°œë³€ìˆ˜: {required}")
    print(f"ê¸°ë³¸ê°’ ë§¤ê°œë³€ìˆ˜: {default}")
    
    if args:
        print(f"ì¶”ê°€ ìœ„ì¹˜ ì¸ì: {args}")
    
    if kwargs:
        print(f"í‚¤ì›Œë“œ ì¸ì: {kwargs}")
    
    print("-" * 40)

# ë‹¤ì–‘í•œ í˜¸ì¶œ
super_function("required_value")
print()

super_function("required_value", "custom_default", 1, 2, 3)
print()

super_function(
    "required_value",
    "custom_default",
    1, 2, 3,
    name="Python",
    version=3.9
)
```

## 4ï¸âƒ£ ì–¸íŒ¨í‚¹(Unpacking) ì—°ì‚°ì

```python
# ğŸ® ì‹¤ìŠµ: * ì—°ì‚°ìë¡œ ì–¸íŒ¨í‚¹

def add_three_numbers(a, b, c):
    return a + b + c

# ì¼ë°˜ì ì¸ í˜¸ì¶œ
result1 = add_three_numbers(1, 2, 3)
print(f"ì¼ë°˜ í˜¸ì¶œ: {result1}")

# ë¦¬ìŠ¤íŠ¸/íŠœí”Œ ì–¸íŒ¨í‚¹
numbers = [4, 5, 6]
result2 = add_three_numbers(*numbers)  # * ì‚¬ìš©!
print(f"ë¦¬ìŠ¤íŠ¸ ì–¸íŒ¨í‚¹: {result2}")

# ë”•ì…”ë„ˆë¦¬ ì–¸íŒ¨í‚¹
def greet(name, age, city):
    print(f"{name}({age}), {city}")

info = {'name': 'ì² ìˆ˜', 'age': 25, 'city': 'ì„œìš¸'}
greet(**info)  # ** ì‚¬ìš©!
```

```python
# ğŸ® ì‹¤ìŠµ: print() í•¨ìˆ˜ì˜ ë¹„ë°€

# print í•¨ìˆ˜ë„ *argsë¥¼ ì‚¬ìš©!
print(1, 2, 3, 4, 5)  # ì—¬ëŸ¬ ê°œ ì¸ì

# ë‚´ë¶€ì ìœ¼ë¡œ ì´ë ‡ê²Œ ë™ì‘
def my_print(*args, sep=' ', end='\n'):
    result = sep.join(str(arg) for arg in args)
    print(result, end=end)

my_print("Hello", "World", "Python")
my_print("A", "B", "C", sep='-')
```

## 5ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: ë¡œê·¸ í•¨ìˆ˜

```python
# ğŸ® ì‹¤ìŠµ: ìœ ì—°í•œ ë¡œê·¸ í•¨ìˆ˜
import datetime

def log(message, *args, level="INFO", **kwargs):
    """
    ìœ ì—°í•œ ë¡œê·¸ í•¨ìˆ˜
    
    Args:
        message: ë¡œê·¸ ë©”ì‹œì§€
        *args: ì¶”ê°€ ë©”ì‹œì§€
        level: ë¡œê·¸ ë ˆë²¨
        **kwargs: ì¶”ê°€ ì •ë³´
    """
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # ë ˆë²¨ë³„ ì´ëª¨ì§€
    level_emoji = {
        "DEBUG": "ğŸ”",
        "INFO": "â„¹ï¸",
        "WARNING": "âš ï¸",
        "ERROR": "âŒ",
        "CRITICAL": "ğŸ”¥"
    }
    
    emoji = level_emoji.get(level, "ğŸ“")
    
    # ê¸°ë³¸ ë¡œê·¸ ì¶œë ¥
    print(f"{emoji} [{timestamp}] [{level}] {message}")
    
    # ì¶”ê°€ ë©”ì‹œì§€
    if args:
        for arg in args:
            print(f"    {arg}")
    
    # ì¶”ê°€ ì •ë³´
    if kwargs:
        print("    Details:")
        for key, value in kwargs.items():
            print(f"      {key}: {value}")

# ì‚¬ìš© ì˜ˆì‹œ
log("ì• í”Œë¦¬ì¼€ì´ì…˜ ì‹œì‘")

log("\nì‚¬ìš©ì ë¡œê·¸ì¸", 
    "ë¡œê·¸ì¸ ì„±ê³µ",
    level="INFO",
    user="admin",
    ip="192.168.1.1")

log("\nì—ëŸ¬ ë°œìƒ",
    "데ì´í„°ë² ì´ìŠ¤ ì—°ê²° ì‹¤íŒ¨",
    "ì¬ì‹œë„ í•„ìš”",
    level="ERROR",
    error_code=500,
    database="main_db")
```

## 7ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: ë°ì½”ë ˆì´í„° íŒ¨í„´

```python
# ğŸ® ì‹¤ìŠµ: ë°ì½”ë ˆì´í„° í•¨ìˆ˜ ë§Œë“¤ê¸°

def debug_decorator(func):
    """í•¨ìˆ˜ í˜¸ì¶œì„ ë””ë²„ê·¸í•˜ëŠ” ë°ì½”ë ˆì´í„°"""
    def wrapper(*args, **kwargs):
        print(f"ğŸ” í•¨ìˆ˜ í˜¸ì¶œ: {func.__name__}")
        print(f"  ìœ„ì¹˜ ì¸ì: {args}")
        print(f"  í‚¤ì›Œë“œ ì¸ì: {kwargs}")
        
        result = func(*args, **kwargs)  # ì›ë˜ í•¨ìˆ˜ í˜¸ì¶œ
        
        print(f"  ë°˜í™˜ê°’: {result}")
        print("-" * 40)
        return result
    
    return wrapper

# ë°ì½”ë ˆì´í„° ì‚¬ìš©
@debug_decorator
def add(a, b):
    return a + b

@debug_decorator
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# í…ŒìŠ¤íŠ¸
result1 = add(3, 5)
result2 = greet("Python", greeting="Hi")
```

## 8ï¸âƒ£ ì‹¤ì „ ì˜ˆì œ: ì„¤ì • ê°€ëŠ¥í•œ í´ë˜ìŠ¤

```python
# ğŸ® ì‹¤ìŠµ: ìœ ì—°í•œ í´ë˜ìŠ¤ ìƒì„±

class FlexibleConfig:
    """ìœ ì—°í•œ ì„¤ì • í´ë˜ìŠ¤"""
    
    def __init__(self, name, **settings):
        self.name = name
        self.settings = settings
        
        # ê¸°ë³¸ ì„¤ì •
        self.defaults = {
            'debug': False,
            'timeout': 30,
            'retries': 3,
            'encoding': 'utf-8'
        }
        
        # ê¸°ë³¸ê°’ê³¼ ì‚¬ìš©ì ì„¤ì • í•©ì¹˜ê¸°
        self.config = {**self.defaults, **settings}
    
    def get(self, key, default=None):
        return self.config.get(key, default)
    
    def update(self, **new_settings):
        self.config.update(new_settings)
        print(f"âœ… {self.name} ì„¤ì • ì—…ë°ì´íŠ¸")
    
    def display(self):
        print(f"ğŸ“‹ {self.name} ì„¤ì •")
        print("-" * 30)
        for key, value in self.config.items():
            print(f"  {key}: {value}")
        print()

# ì‚¬ìš© ì˜ˆì‹œ
# ê¸°ë³¸ ì„¤ì •ë§Œ
config1 = FlexibleConfig("ê°œë°œ í™˜ê²½")
config1.display()

# ì»¤ìŠ¤í…€ ì„¤ì •
config2 = FlexibleConfig(
    "í”„ë¡œë•ì…˜ í™˜ê²½",
    debug=True,
    timeout=60,
    database="postgresql",
    cache=True,
    max_connections=100
)
config2.display()

# ì„¤ì • ì—…ë°ì´íŠ¸
config2.update(cache=False, timeout=120)
config2.display()
```
