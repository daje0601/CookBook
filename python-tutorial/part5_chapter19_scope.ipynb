## 4ï¸âƒ£ ì¤‘ì²© í•¨ìˆ˜ì™€ nonlocal

```python
# ğŸ® ì‹¤ìŠµ: í•¨ìˆ˜ ì•ˆì˜ í•¨ìˆ˜

def outer_function():
    message = "ì™¸ë¶€ í•¨ìˆ˜ì˜ ë³€ìˆ˜"  # outerì˜ ì§€ì—­ ë³€ìˆ˜
    
    def inner_function():
        # innerì—ì„œ outerì˜ ë³€ìˆ˜ ì ‘ê·¼ ê°€ëŠ¥!
        print(f"ë‚´ë¶€ í•¨ìˆ˜ì—ì„œ: {message}")
    
    print(f"ì™¸ë¶€ í•¨ìˆ˜ì—ì„œ: {message}")
    inner_function()

outer_function()
```

```python
# ğŸ® ì‹¤ìŠµ: nonlocal í‚¤ì›Œë“œ

def counter_factory():
    count = 0  # ì™¸ë¶€ í•¨ìˆ˜ì˜ ì§€ì—­ ë³€ìˆ˜
    
    def increment():
        nonlocal count  # ì™¸ë¶€ í•¨ìˆ˜ì˜ ë³€ìˆ˜ ì‚¬ìš© ì„ ì–¸
        count += 1
        return count
    
    def decrement():
        nonlocal count
        count -= 1
        return count
    
    def get_count():
        return count  # ì½ê¸°ë§Œ í•˜ë©´ nonlocal ë¶ˆí•„ìš”
    
    # ë‚´ë¶€ í•¨ìˆ˜ë“¤ì„ ë°˜í™˜
    return increment, decrement, get_count

# ì‚¬ìš©í•˜ê¸°
inc, dec, get = counter_factory()

print(f"ì´ˆê¸°ê°’: {get()}")
print(f"ì¦ê°€: {inc()}")
print(f"ì¦ê°€: {inc()}")
print(f"ê°ì†Œ: {dec()}")
print(f"í˜„ì¬ê°’: {get()}")
```

## 5ï¸âƒ£ LEGB ê·œì¹™

```python
# ğŸ® ì‹¤ìŠµ: Pythonì˜ ë³€ìˆ˜ íƒìƒ‰ ìˆœì„œ (LEGB)
# L: Local (ì§€ì—­)
# E: Enclosing (ë‘˜ëŸ¬ì‹¼ í•¨ìˆ˜)
# G: Global (ì „ì—­)
# B: Built-in (ë‚´ì¥)

# Built-in (ë‚´ì¥ í•¨ìˆ˜)
print("lenì€ ë‚´ì¥ í•¨ìˆ˜:", len([1, 2, 3]))

# Global (ì „ì—­)
x = "ì „ì—­ x"

def outer():
    # Enclosing (ë‘˜ëŸ¬ì‹¼)
    x = "ì™¸ë¶€ í•¨ìˆ˜ì˜ x"
    
    def inner():
        # Local (ì§€ì—­)
        x = "ë‚´ë¶€ í•¨ìˆ˜ì˜ x"
        print(f"1. innerì—ì„œ x = {x}")  # Local x ì‚¬ìš©
    
    inner()
    print(f"2. outerì—ì„œ x = {x}")  # Enclosing x ì‚¬ìš©

outer()
print(f"3. ì „ì—­ì—ì„œ x = {x}")  # Global x ì‚¬ìš©
```

```python
# ğŸ® ì‹¤ìŠµ: í´ë¡œì €ë¥¼ ì´ìš©í•œ í”„ë¼ì´ë¹— ë³€ìˆ˜

def make_bank_account(initial_balance):
    """ì€í–‰ ê³„ì¢Œ ìƒì„± (í´ë¡œì € í™œìš©)"""
    balance = initial_balance  # ì™¸ë¶€ì—ì„œ ì§ì ‘ ì ‘ê·¼ ë¶ˆê°€!
    
    def deposit(amount):
        nonlocal balance
        if amount > 0:
            balance += amount
            print(f"ğŸ’µ {amount:,}ì› ì…ê¸ˆ")
            return balance
        return None
    
    def withdraw(amount):
        nonlocal balance
        if 0 < amount <= balance:
            balance -= amount
            print(f"ğŸ’¸ {amount:,}ì› ì¶œê¸ˆ")
            return balance
        else:
            print("âŒ ì”ì•¡ ë¶€ì¡±!")
            return None
    
    def get_balance():
        return balance
    
    # í•¨ìˆ˜ë“¤ì„ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜
    return {
        'deposit': deposit,
        'withdraw': withdraw,
        'balance': get_balance
    }

# ê³„ì¢Œ ìƒì„±
account = make_bank_account(10000)

print(f"ì´ˆê¸° ì”ì•¡: {account['balance']():,}ì›")
account['deposit'](5000)
account['withdraw'](3000)
account['withdraw'](20000)  # ì”ì•¡ ë¶€ì¡±
print(f"ìµœì¢… ì”ì•¡: {account['balance']():,}ì›")

# balance ë³€ìˆ˜ì— ì§ì ‘ ì ‘ê·¼ ë¶ˆê°€!
# print(balance)  # âŒ NameError
```
